```python
def generate_modular_fc(num_modules, nodes_per_module):
    # Generate a modular functional connectivity matrix
    size = num_modules * nodes_per_module
    fc_matrix = np.zeros((size, size))

    for i in range(num_modules):
        start_row = i * nodes_per_module
        end_row = start_row + nodes_per_module
        for j in range(num_modules):
            if i != j:
                start_col = j * nodes_per_module
                end_col = start_col + nodes_per_module
                fc_matrix[start_row:end_row, start_col:end_col] = np.random.rand(nodes_per_module, nodes_per_module)

    return fc_matrix + fc_matrix.T
```

```python
import numpy as np

def generate_weighted_sbm(num_nodes, cluster_sizes):
    num_clusters = len(cluster_sizes)
    adjacency_matrix = np.zeros((num_nodes, num_nodes))

    # Assign nodes to clusters
    nodes = range(num_nodes)
    start = 0
    clusters = []
    for i, size in enumerate(cluster_sizes):
        clusters += [i]*size
        

    # Generate weights for within-cluster and between-cluster connections
    within_cluster_weight = 0.8
    between_cluster_weight = 0.1

    for i in range(num_nodes):
        for j in range(i + 1, num_nodes):

            if clusters[i] == clusters[j]:
                weight = np.random.normal(within_cluster_weight, 0.05)  # Adding some noise
                adjacency_matrix[i, j] = weight
                adjacency_matrix[j, i] = weight
            else:
                weight = np.random.normal(between_cluster_weight, 0.05)  # Adding some noise
                adjacency_matrix[i, j] = weight
                adjacency_matrix[j, i] = weight
                    
    print(nodes)

    return adjacency_matrix

# Example usage:
num_nodes = 50
cluster_sizes = [10]*5  # Example cluster sizes

adj_matrix = generate_weighted_sbm(num_nodes, cluster_sizes)
#print(adj_matrix)  # Output the generated adjacency matrix

dist_mat = 1 - adj_matrix
```

## Multiple realisations

```python
N = 300 # number of points
num_it = 50 # number of iterations
num_parents = [10, 20, 30] # numbers of clusters
radii = [0.05, 0.1, 0.3, 0.5] # radii around each cluster

# uniform point process
samples_uniform = PoissonPointProcess(window, N, num_it, conditioning=True, asPP=False)

# clustered point process
samples_clustered = {}
for num_par in tqdm(num_parents):
    samples_clustered[num_par] = {}
    for radius in radii:
        samples_clustered[num_par][radius] = PoissonClusterPointProcess(
            window, N, num_par, radius, num_it, asPP=False, conditioning=True
        )
```

```python
# compute diagrams for uniform
topo_uniforms = []
for _, sample in samples_uniform.realizations.items():
    pp = PointPattern(sample).df
    diag = rip.ripser(pp)
    topo_uniforms.append(diag)
```

```python
# compute diagrams for clustered
topo_clustereds = {}
for num_par in tqdm(num_parents):  # number clusters
    topo_clustereds[num_par] = {}
    for radius in tqdm(radii):  # radius
        topo_clustereds[num_par][radius] = []
        for _, sample in samples_clustered[num_par][
            radius
        ].realizations.items():  # realization
            pp = PointPattern(sample).df
            diag = rip.ripser(pp)
            topo_clustereds[num_par][radius].append(diag)
```

```python
def tot_persistence(diag):
    return np.sum(np.diff(diag[:-1]))
```

```python
# compute topo descriptos
dims_homology = [0, 1]

entropy_uniform = defaultdict(list)
tot_persistence_uniform = defaultdict(list)

for dgms in topo_uniforms:
    diag = dgms["dgms"]
    es = persistent_entropy.persistent_entropy(diag)
    
    for dim in dims_homology:
        entropy_uniform[dim].append(es[dim])
        tot_persistence_uniform[dim].append(tot_persistence(diag[dim]))
```

```python
def move_sn_y(offs=0, dig=0, side='left', omit_last=False):
    """Move scientific notation exponent from top to the side.
    
    Additionally, one can set the number of digits after the comma
    for the y-ticks, hence if it should state 1, 1.0, 1.00 and so forth.

    Parameters
    ----------
    offs : float, optional; <0>
        Horizontal movement additional to default.
    dig : int, optional; <0>
        Number of decimals after the comma.
    side : string, optional; {<'left'>, 'right'}
        To choose the side of the y-axis notation.
    omit_last : bool, optional; <False>
        If True, the top y-axis-label is omitted.

    Returns
    -------
    locs : list
        List of y-tick locations.

    Note
    ----
    This is kind of a non-satisfying hack, which should be handled more
    properly. But it works. Functions to look at for a better implementation:
    ax.ticklabel_format
    ax.yaxis.major.formatter.set_offset_string
    """

    # Get the ticks
    locs, _ = plt.yticks()

    # Put the last entry into a string, ensuring it is in scientific notation
    # E.g: 123456789 => '1.235e+08'
    llocs = '%.3e' % locs[-1]

    # Get the magnitude, hence the number after the 'e'
    # E.g: '1.235e+08' => 8
    yoff = int(str(llocs).split('e')[1])

    # If omit_last, remove last entry
    if omit_last:
        slocs = locs[:-1]
    else:
        slocs = locs

    # Set ticks to the requested precision
    form = r'$%.'+str(dig)+'f$'
    plt.yticks(locs, list(map(lambda x: form % x, slocs/(10**yoff))))

    # Define offset depending on the side
    if side == 'left':
        offs = -.18 - offs # Default left: -0.18
    elif side == 'right':
        offs = 1 + offs    # Default right: 1.0
        
    # Plot the exponent
    plt.text(offs, .98, r'$\times10^{%i}$' % yoff, transform =
            plt.gca().transAxes, verticalalignment='top')

    # Return the locs
    return locs
```
