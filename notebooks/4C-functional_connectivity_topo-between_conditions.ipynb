# Functional connectivity: topo differences between Highs and Lows

```python
# import necessary libraries
import pickle as pk
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
from rich import pretty
from sklearn.metrics import pairwise_distances
from tqdm import tqdm

pretty.install()

sb.set_theme(style="ticks", context="paper")

results_dir = "../results/functional_connectivity_topo/"
data_dir = "../data/functional_connectivity_topo/"
data_out_dir = "../data/functional_connectivity_topo/"
Path(results_dir).mkdir(parents=True, exist_ok=True)
Path(data_dir).mkdir(parents=True, exist_ok=True)
Path(data_out_dir).mkdir(parents=True, exist_ok=True)
```

## Load data

```python
skip = 1  # factor by which the input data is to be downsampled
n_windows = 18  # per condition

# topo_descriptors_labeled = pd.read_pickle("../data/corrtopodescriptorslabeled.pkl")
topo_descriptors_labeled = pk.load(
    open(f"{data_dir}topo_descriptors_labeled_{n_windows}wind_skip_{skip}.pkl", "rb")
)
```

```python
max_dim = 1  # maximum homology dimension
```

## Compare topological descriptors between conditions

```python
# we compute the distance between each pair of persistence diagrams
distance_matrices = {}
dist_long = {}

for dim in range(max_dim + 1):
    #var = ["tot_persistence", "persistent_entropy", "betti"]
    var = ["avg_persistence"]#, "persistent_entropy"]  # , "persistent_entropy"]
    suffix = "-".join(var)
    topos = topo_descriptors_labeled[dim][var]

    distance_matrix = pairwise_distances(topos, metric="euclidean")

    distance_matrices[dim] = pd.DataFrame(
        distance_matrix,
        columns=topos.index,
        index=topos.index,
    )

    # change to long format for easier plotting and filtering
    dist_long[dim] = distance_matrices[dim].stack().reset_index()
    dist_long[dim].columns = ["series_i", "series_j", "distance"]
    dist_long[dim] = dist_long[dim][["distance", "series_i", "series_j"]]

    # mask redundant lines, when series pairs are considered twice (0, 1) and (1, 0)
    # mask_sym_pairs = dist_long[dim]["series_i"] >= dist_long[dim]["series_j"]
    dist_long[dim] = dist_long[dim]  # [mask_sym_pairs]
```

```python
cols = ["file", "group", "subject", "condition"]

df_to_merge = topo_descriptors_labeled[dim][cols]
df_to_merge

dist_long_labeled = dict()

for dim in range(max_dim + 1):
    dist_long_labeled_i = dist_long[dim].merge(
        right=df_to_merge,
        left_on="series_i",
        right_index=True,
        how="left",
        suffixes=["_i", "_j"],
    )
    dist_long_labeled[dim] = dist_long_labeled_i.merge(
        right=df_to_merge,
        left_on="series_j",
        right_index=True,
        how="left",
        suffixes=["_i", "_j"],
    )
    # Aggiungi la colonna "Conditions"
    dist_long_labeled[dim]["conditions"] = (
        dist_long_labeled[dim]["condition_i"].astype(str)
        + "-"
        + dist_long_labeled[dim]["condition_j"].astype(str)
    )
    dist_long_labeled[dim]["groups"] = (
        dist_long_labeled[dim]["group_i"].astype(str)
        + "-"
        + dist_long_labeled[dim]["group_j"].astype(str)
    )
```

```python
# avoid counting doubles
conds_ordered = dist_long_labeled[dim]["condition_i"].unique().tolist()

for dim in range(max_dim + 1):

    mask_double_subj = (
        dist_long_labeled[dim]["subject_i"] >= dist_long_labeled[dim]["subject_j"]
    )
    mask_same_subj = (
        dist_long_labeled[dim]["subject_i"] == dist_long_labeled[dim]["subject_j"]
    )

    # cond_i_idx = dist_long_labeled[dim]["condition_i"].apply(lambda x: conds_ordered.index(x))
    # cond_j_idx = dist_long_labeled[dim]["condition_j"].apply(lambda x: conds_ordered.index(x))
    # mask_double_cond = cond_i_idx >= cond_j_idx # needs index because alphabetical order is not temporal

    # dist_long_labeled[dim] = dist_long_labeled[dim][mask_double] # windows are still double
    # window_i = dist_long_labeled[dim]["file_i"].str.rsplit("_").str[-1]
    # window_j = dist_long_labeled[dim]["file_j"].str.rsplit("_").str[-1]
    # mask_double_windows = window_i >= window_j

    # mask_double = mask_double_subj * mask_double_windows
    mask_same_group = (
        dist_long_labeled[dim]["group_i"] == dist_long_labeled[dim]["group_j"]
    )

    mask = mask_same_subj * mask_same_group

    dist_long_labeled[dim] = dist_long_labeled[dim][mask]
```

## Effect size between Highs and Lows

```python
from pingouin import compute_effsize
```

```python
effect_size = {}
n_cond = 6

conditions_list = ["ROE", "RCE", "IND1", "IND2", "NH", "POST"]

for dim in range(max_dim + 1):

    mask_high = (dist_long_labeled[dim]["group_i"] == "high") & (
        dist_long_labeled[dim]["group_j"] == "high"
    )
    mask_low = (dist_long_labeled[dim]["group_i"] == "low") & (
        dist_long_labeled[dim]["group_j"] == "low"
    )

    effect_size_k = np.zeros((n_cond, n_cond))

    for i, condition_i in enumerate(conditions_list):
        mask_i = dist_long_labeled[dim]["condition_i"] == condition_i

        for j, condition_j in enumerate(conditions_list):
            mask_j = dist_long_labeled[dim]["condition_j"] == condition_j

            mask = mask_i & mask_j
            mask1 = mask_i & mask_j & mask_high
            mask2 = mask_i & mask_j & mask_low

            dist_filtered_high = dist_long_labeled[dim][mask1]
            dist_filtered_low = dist_long_labeled[dim][mask2]

            high_mean = dist_filtered_high["distance"].mean()
            low_mean = dist_filtered_low["distance"].mean()

            high_std = dist_filtered_high["distance"].std()
            low_std = dist_filtered_low["distance"].std()

            n_high = len(dist_filtered_high["distance"])
            n_low = len(dist_filtered_low["distance"])

            difference = high_mean - low_mean
            pooled_std = np.sqrt(
                ((high_std**2 * (n_high - 1)) + (low_std**2 * (n_low - 1)))
                / (n_high + n_low - 2)
            )

            effect_size_k[i, j] = compute_effsize(
                dist_filtered_high["distance"], dist_filtered_low["distance"]
            )  # difference / pooled_std
    effect_size[dim] = effect_size_k
```

```python
conditions_list = ["ROE", "RCE", "IND1", "IND2", "NH", "POST"]

fig, axs = plt.subplots(1, 2, figsize=(6, 4), sharey=True, layout="constrained")

for dim in range(max_dim + 1):

    ax = axs[dim]

    # effect_size_masked = np.where(mask_stats[dim], np.nan, effect_size[dim])

    title = f"$H_{dim}$"
    ax.set_title(title)

    ax = sb.heatmap(
        data=effect_size[dim],
        cmap="coolwarm",
        vmin=-0.3,
        vmax=0.3,
        square=True,
        annot=True,
        fmt=".2f",
        center=0,
        xticklabels=conditions_list,
        yticklabels=conditions_list,
        ax=ax,
        cbar=dim == 1,
        cbar_kws={"shrink": 0.5},
    )

    # print(np.nanmax(effect_size_masked))
    # print(np.nanmin(effect_size_masked))

# fig.colorbar(ax.get_children()[0], ax=axs[-1])

fig.suptitle("FC topology: effect size High-Low")

plt.show()
```

## Barplots of topo distances in each condition pair

```python
dist_filtered = {}

for dim in range(max_dim + 1):
    mask_same_group = (
        dist_long_labeled[dim]["group_i"] == dist_long_labeled[dim]["group_j"]
    )
    # mask_same_subject = dist_long_labeled[dim]["subject_i"] == dist_long_labeled[dim]["subject_j"]
    # mask_same_conditions = dist_long_labeled[dim]["condition_i"] == dist_long_labeled[dim]["condition_j"]

    mask = mask_same_group

    dist_filtered[dim] = dist_long_labeled[dim][mask]
```

```python
for dim in range(max_dim + 1):

    fig, ax = plt.subplots(1, 1, figsize=(8, 2.5), sharex=True, sharey=True)

    title = f"FC topology: distance between conditions, $H_{dim}$"
    plt.suptitle(title)

    sb.barplot(
        x="conditions",
        y="distance",
        data=dist_filtered[dim],
        ax=ax,
        hue="groups",
        # order=unique_pairs_df["conditions"],
    )

    plt.xticks(rotation=90)
    sb.despine()

    var = dist_filtered[dim]["conditions"].unique()
    pairs4 = [[(i, "high-high"), (i, "low-low")] for i in var]
    pairs = pairs4

    from statannotations.Annotator import Annotator

    # mask = dist_filtered[dim]["conditions"] == var[i]
    annotator = Annotator(
        ax,
        pairs,
        data=dist_filtered[dim],
        x="conditions",
        y="distance",
        hue="groups",
        verbose=False
        # order=order,
        # hue_order=["high-high", "low-low"],
    )
    annotator.configure(test="Wilcoxon", text_format="star", loc="inside")
    ax, stat_results[dim] = annotator.apply_and_annotate()

    plt.tight_layout()

    fig_name = f"topo_features_distances_barplots_H{dim}_{n_windows}wind_skip_{skip}_{suffix}"
    plt.savefig(f"{results_dir}{fig_name}.png", facecolor="w", dpi=250, bbox_inches="tight")

    # plt.savefig(f"{figname}.png", facecolor="white")
```

## Mask effect size based on stats

```python
mask_stats = {}
stars_stats = {}

for dim in range(max_dim + 1):

    stat_conds = [i.data.group1[0].split("-") for i in stat_results[dim]]
    stat_conds_ids = [
        (conditions_list.index(i), conditions_list.index(j)) for i, j in stat_conds
    ]

    # mask_stats_list = np.array([i.text for i in stat_results[dim]]) == "ns"
    stars_stats_list = [i.text for i in stat_results[dim]]

    stars_arr = np.zeros((6, 6), dtype=object)
    mask_arr = np.zeros((6, 6), dtype=bool)

    for idx, val in zip(stat_conds_ids, stars_stats_list):
        mask_arr[idx] = True if val == "ns" else False
        stars_arr[idx] = val

    mask_stats[dim] = mask_arr
    stars_stats[dim] = stars_arr
```

```python
conditions_list = ["ROE", "RCE", "IND1", "IND2", "NH", "POST"]

fig, axs = plt.subplots(1, 2, figsize=(5, 3), sharey=True, layout="constrained")

for dim in range(max_dim + 1):

    ax = axs[dim]

    effect_size_masked = np.where(mask_stats[dim], np.nan, effect_size[dim])

    title = f"$H_{dim}$"
    ax.set_title(title)

    ax = sb.heatmap(
        data=effect_size_masked,
        cmap="coolwarm",
        vmin=-0.3,
        vmax=0.3,
        square=True,
        annot=stars_stats[dim],
        fmt="",
        center=0,
        xticklabels=conditions_list,
        yticklabels=conditions_list,
        ax=ax,
        cbar=dim == 1,
        cbar_kws={"shrink": 0.5},
    )

    # print(np.nanmax(effect_size_masked))
    # print(np.nanmin(effect_size_masked))

# fig.colorbar(ax.get_children()[0], ax=axs[-1])

fig.suptitle(f"FC {suffix}: effect size High-Low")

fig_name = f"topo_features_effect_size_filtered_{n_windows}wind_skip_{skip}_{suffix}"
plt.savefig(f"{results_dir}{fig_name}.png", facecolor="w", dpi=250, bbox_inches="tight")

plt.show()
```
